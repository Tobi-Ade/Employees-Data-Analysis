```python
def categorical_columns(employee_data):
    """
    This function takes in the dataframe and filters out the columns with object data type
    :params: pandas dataframe 
    :return: object type columns 
    :rtype: dataframe object
    """
    categorical = employee_data.select_dtypes(include="object")
    return categorical
```

```python
categorical_columns(employee_data)
```

```python
def categorical_unique(employee_data):
    """
    This function takes in a dataframe and returns a list of the unique values in all columns with data type 'object'
    :params: pandas dataframe 
    :return: Unique values in object type columns 
    :rtype: List 
    """
    unique_attributes = [[column, employee_data[column].unique()] for column in employee_data if employee_data[column].dtype=="object"]
    return unique_attributes
```

```python
categorical_unique(employee_data)
```

```python
def numerical_columns(employee_data):
    """
    This function takes in the dataframe and filter columns with numerical values
    :params: pandas dataframe
    :return: numerical columns dataframe, numerical columns names 
    :rtype: dataframe object, pandas index
    """
    numerical = employee_data.select_dtypes(include=['float64', 'int64'])
    numerical_cols = numerical.columns
    return numerical, numerical_cols
```

```python
def numerical_unique(employee_data):
    """
    This function takes a dataframe and returns the unique values in each of the specified columns
    :params: pandas dataframe 
    :return: Unique values in specified columns
    :rtype: List
        
    """
    #Specifying the columns to be used in the function
    columns = ['Education', "EnvironmentSatisfaction", "JobInvolvement", "JobLevel", "JobSatisfaction", "PerformanceRating", 
    "RelationshipSatisfaction", "StockOptionLevel", "WorkLifeBalance"]
    numerical = employee_data[columns]
    unique_numerical = [[column, numerical[column].unique()] for column in numerical]
    return unique_numerical
```

                                            FILLING MISSING VALUES 

```python
def missing_data_percent(employee_data):
    """
    This function takes in a dataframe and returns the sum and percentage of missing data in all the dataframe columns
    :params: pandas dataframe 
    :return: sum and percentage of missing data in all columns 
    :rtype: dataframe object     
    """
    missing_total = employee_data.isnull().sum()
    len_data = len(employee_data)
    missing_percent = round((missing_total/len_data)*100, 3)
    missing_data = pd.concat([missing_total, missing_percent], axis=1, keys = ['Total Missing Data', '% Missing Data'])
    return missing_data
```

```python
missing_data_percent(employee_data)
```

```python
"""
Creating a copy of the dataset to be used for analysis 
"""
employee_data_copy = employee_data.copy()
```

```python
def sum_missing_values(employee_data):
    """
    This function takes a dataframe and returns the columns that have missing values as well as the number of missing values
    :params: pandas dataframe
    :return: columns with missing values and the number of missing values 
    :rtype: List
    """
    missing_values = [[column, employee_data_copy[column].isna().sum().sum()] for column in employee_data.columns if employee_data[column].isna().sum().sum()]
    return missing_values
```

```python
sum_missing_values(employee_data_copy)
```

We can see the 5 columns with missing data and the number of entries that are missing. We proceed to handle each of them one after the other 

```python
"""
Filling the missing values in the DailyRate column 
The code line is displaying a slice of the dataframe where the DailyRate is missing 
"""
employee_data_copy[employee_data['DailyRate'].isna()]
```

```python
"""
Printing out the mean DailyRate of the departments where the DailyRate is missing
To fill the data with the mean, we have to check if the total mean and the mean for each department is not far off in valuation
The result shows that between them, there is a difference of just more than 6 which is acceptable 
"""
print("The mean daily rate in R&D is", round(employee_data_copy[employee_data_copy['Department']=="Research & Development"]['DailyRate'].mean(), 3))
print("The mean daily rate in Sales is", round(employee_data_copy[employee_data_copy['Department']=="Sales"]['DailyRate'].mean(), 3))
print("The mean daily rate of all employees regardless of department is", round(employee_data['DailyRate'].mean(), 3))
```

```python
"""
Filling the missing values with the mean makes sure the fill value is within range of the DailyRate distribution
This is done with the fillna method 
setting inplace to True ensures that the fill is permanent
"""
employee_data_copy['DailyRate'].fillna(round(employee_data_copy["DailyRate"].mean(), 1), inplace=True)
```

```python
"""
Displaying the row where the DistanceFromHome is missing 
"""
employee_data_copy[employee_data_copy['DistanceFromHome'].isna()]
```

```python
"""
Filling the missing distance with the mean
"""
employee_data_copy['DistanceFromHome'].fillna(employee_data['DistanceFromHome'].mean(), inplace=True)
```

```python
"""
Displaying the rows where the Education is missing
"""
employee_data_copy.loc[employee_data_copy.Education.isna()]
```

```python
def mean_department_education(department_name):
    """
    This function takes in a variable name which must be one of the departments in the dataset
    It returns the mean Education for that depaartment
    :params: string which must be a department name 
    :return: The department's mean Education
    :rtype: float
    """
    department = employee_data_copy[employee_data_copy['Department'] == department_name]
    return department['Education'].mean()

def median_department_education(department_name):
    """
    This function takes in a variable name which must be one of the departments in the dataset
    It returns the median Education for that depaartment
    :params: string which must be a department name 
    :return: The department's median Education
    :rtype: float
    """
    department = employee_data_copy[employee_data_copy['Department'] == department_name]
    return department['Education'].median()

```

```python
"""
Checking the mean Education per department
We can see that the mean Education of all the departments when rounded is very close to 3.0
All the acceptable Education values are whole numbers which range from 1-4 
Therefore it makes sense to fill the missing Education with 3
"""
print("The mean education of Sales is ", mean_department_education("Sales"))
print("The mean education of Research & Development is ", mean_department_education("Research & Development"))
print("The mean education of Human Resources is ", mean_department_education("Human Resources"))
print("The mean education of the employees regardless of department is ", employee_data["Education"].mean())
```

```python
""""
Filling the missing Education values with 3
"""
employee_data_copy['Education'].fillna(3.0, inplace=True)
```

```python
"""
Displaying the rows of the dataset where the Hourly Rate is Missing 
The result shows that the missing Hourly Rate are all employees in R&D department, except one 
"""
employee_data_copy.loc[employee_data_copy.HourlyRate.isna()]
```

```python
"""
Checking the mean, median and mode of the Hourly Rate for R&D
"""
print("The mean HourlyRate of employees in R&D is", employee_data_copy.loc[employee_data.Department=='Research & Development','HourlyRate'].mean())

print("The median HourlyRate of employees in R&D is", employee_data_copy.loc[employee_data.Department=='Research & Development','HourlyRate'].median())

print("The mode HourlyRate of employees in R&D is", employee_data_copy.loc[employee_data.Department=='Research & Development','HourlyRate'].mode())

print("The mean HourlyRate of all employees is", round(employee_data_copy["HourlyRate"].mean(),0))

```


The result shows that the mean and the mode are similar<br>
The mean HourlyRate regardless of department is also in the same range<br>
Therefore the mode is suitable for filling the missing HourlyRate

```python
"""
Filling the missing Hourly Rate with the mean rounded up to an acceptable value
"""
employee_data_copy['HourlyRate'].fillna(round(employee_data_copy["HourlyRate"].mean(),0), inplace=True)
```

```python
"""
Checking the number of missing Job Involvements per department 
The result shows that most of the missing Job Involvements are in Research & Development
"""
employee_data_copy.loc[employee_data_copy.JobInvolvement.isna(),'Department'].value_counts()
```

```python
"""
Checking the count of Job involvement 
The result shows that the highest level of Job Involvement in 3.0 by a wide margin
Therefore it makes to fill the missing Job Involvement with the mode 
"""
employee_data['JobInvolvement'].value_counts()
```

```python
"""
Filling the missing Job Involvements with the mode 
"""
employee_data_copy['JobInvolvement'].fillna(3.0, inplace=True)
```

```python
"""
Calling the function that checks the sum of missing values in the dataset
The result shows that EmployeeNumber has 4 missing values
The column will be dropped in the preprocessing stage
By doing this, we have taken care of all missing values 
"""
sum_missing_values(employee_data_copy)
```

```python
"""
Dropping single value columns / redundant columns 
"""
employee_data_copy.drop(columns=["EmployeeCount", "EmployeeNumber", "StandardHours", "Over18"], inplace=True)
```

```python
def one_hot_encode(employee_data, column):
    """
    This function takes a dataframe and a column name and returns the dataframe with dummies for the specified column
    :params: pandas dataframe, column name 
    :return: dataframe with dummies for specfied column in params
    :rtype: pandas dataframe
    """
    dummies = pd.get_dummies(employee_data[column], prefix=column, drop_first=True) 
    employee_data = pd.concat([employee_data, dummies], axis=1)
    employee_data = employee_data.drop(column, axis=1)
    return employee_data
```

```python
"""
Binary Encoding categorical binary columns by replacing values with 0 or 1
"""
employee_data_copy["Attrition"] = employee_data_copy["Attrition"].replace({"Yes" : 1, "No" : 0})

employee_data_copy["Gender"] = employee_data_copy["Gender"].replace({"Male" : 1, "Female" : 0})

employee_data_copy["OverTime"] = employee_data_copy["OverTime"].replace({"Yes" : 1, "No" : 0})
```
